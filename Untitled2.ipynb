```python
import pandas as pd

# FAQ data direct code me hi likh do
faq_data = {
    'Question': [
        'How much is the admission fee?',
        'How can I apply for the hostel?',
        'When will exams start?',
        'What is the refund policy?',
        'Where can I find timetable?',
        'Is hostel compulsory?',
        'How to contact university helpdesk?'
    ],
    'Answer': [
        'Admission fee is 5000.',
        'Fill the hostel form online at hostel.university.edu.',
        'Exams will begin in December as per academic calendar.',
        'Refund policy details are available on university website.',
        'Timetable is updated on the student portal.',
        'Hostel is optional for all students.',
        'Contact helpdesk at help@university.edu or call 1800-123-456.'
    ]
}

faq_df = pd.DataFrame(faq_data)
print(faq_df)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def preprocess_spacy(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return ' '.join(tokens)

faq_df['processed_question'] = faq_df['Question'].apply(preprocess_spacy)
print(faq_df[['Question', 'processed_question']])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Vectorize processed questions
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(faq_df['processed_question'])

def get_answer(user_query):
    # spaCy preprocessing user query
    user_query_processed = preprocess_spacy(user_query)
    # Transform user query
    user_tfidf = vectorizer.transform([user_query_processed])
    # Similarity check
    similarities = cosine_similarity(user_tfidf, tfidf_matrix).flatten()
    # Best match index
    best_match_idx = similarities.argmax()
    # Recommended answer
    return faq_df.iloc[best_match_idx]['Answer']

# Example: User query test
user_query = "hostel apply karna hai"
print(get_answer(user_query))
```

```python
while True:
    user_query = input("Type your question (or type 'exit' to quit): ")
    if user_query.lower() == 'exit':
        break
    print("Bot answer:", get_answer(user_query))
```
